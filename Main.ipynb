Credit Card Fraud Detection

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
credit=pd.read_csv('/content/creditcard.csv')
```

```python
credit.head()
```

```python
credit.info()
```

```python
credit.describe()
```

```python
print(credit.std())
```

```python
credit.value_counts('Class')
```

```python
credit.tail()
```

```python
credit.isnull().sum()
```

```python
credit=credit.dropna(how='any')
```

```python
credit.isnull().sum()
```

```python
legit=credit[credit.Class==0]
fraud=credit[credit.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample=legit.sample(n=169)
print(legit_sample.shape)
```

```python
new_credit=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_credit.head()
```

```python
new_credit.value_counts('Class')
```

```python
new_credit.groupby('Class').mean()
```

```python
x=new_credit.drop(columns='Class',axis=1)
y=new_credit.Class
```

```python
print(x)
print(y)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
x=scaler.transform(x)
```

```python
print(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
trainpredict=model.predict(xtrain)
trainaccuracy=accuracy_score(trainpredict,ytrain)
print(trainaccuracy)
```

```python
testpredict=model.predict(xtest)
testaccuracy=accuracy_score(testpredict,ytest)
print(testaccuracy)
```
